```python
temp1 = df['Sales']
temp2 = df['Product_Category'].value_counts()
fig = plt.figure() # start to draw
ax1 = fig.add_subplot(121) # 1 row, 2 columns, No.1 chart
ax2 = fig.add_subplot(122) # 1 row, 2 columns, No.2 chart
temp1.plot(kind='kde', ax=ax1) #kind=line kind=bar kind'kernel density estimation'<histogram 
ax1.set_xlabel('Cost')
ax1.set_ylabel('Amount')
ax1.set_title("Cost chart")
temp2.plot(kind='pie', autopct='%1.2f%%', ax=ax2) #< %1=+ 2f%%
ax2.set_xlabel('Category Distribution')
ax2.set_ylabel('Categroy')
ax2.set_title("Category Distribution")
plt.show()
```

```python
temp3=pd.crosstab(df['Education'],df['Gender'])
temp3.plot(kind='bar',stacked=True,color=['red','blue'],grid='False')
```

# Loan Term Preference

```python
loan_cor=df.groupby(["Gender","Loan_Amount_Term"]).count()
loan_cor
```

```python
plt.rc('figure', figsize=(50, 20))
ax = df.plot(x='Loan_Amount_Term',bins=100, kind='bar', stacked=True, color=['#33CCCC', '#FFCC00'])
plt.xlabel('Date', fontsize=30)
plt.ylabel('Sales percentage', fontsize=30)
plt.title('Sales Percentage of Electric Cars and Gasoline Cars Over Years', fontsize=30)
ax.tick_params(axis='both', which='major', labelsize=20)
ax.yaxis.set_minor_locator(plt.MultipleLocator(5))
```
